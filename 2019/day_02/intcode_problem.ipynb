### Part 1

```python
input_file = open('bixo_input')
intcode = [int(char) for char in input_file.read().split(',')]
input_file.close()

intcode[1] = 12
intcode[2] = 2

def add(a,b):
    return a+b

def mult(a,b):
    return a*b

def halt(a,b):
    print('Intcode Computer Halting: ', intcode[0])
    return 'ERROR'
    
op = dict()
op['1'] = add
op['2'] = mult
op['99'] = halt

def execute(pos, verbose = False):
    
    if verbose:
        print(intcode[pos:pos+4])

    opcode, pos1, pos2, pos3 = intcode[pos:pos+4]
    
    int1 = intcode[pos1]
    int2 = intcode[pos2]
    
    ans = op[str(opcode)](int1, int2)
    
    if verbose:
        print(ans)
    
    if ans == 'ERROR':
        return 'ERROR'
    else:
        intcode[pos3] = ans
        
def get_halt_answer(verbose = False):
    for i in range(0, len(intcode), 4):
        if execute(i, verbose = verbose) == 'ERROR':
            break
    return intcode[0]
```
